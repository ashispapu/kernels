# Basic Visualisations using matplotlib

The role of the data scientist and the machine learning engineers do not stop with the creation of the models. They also need to define the predictions of the models and why the model is behaving the way it is. Sometimes they need to convey the shape of the day and the unique properties of the data. There are various data visualisation techniques in python that are very good in making sense of your data and coming up with good visualisations.

### Libraries

There are two main libraries in python that help you to create good data visualisations. They are

1. matplotlib
1. seaborn

Import the dataset

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
%matplotlib inline
data = load_iris()
```

```python
X_sepal_length = data.data[:,0]
X_sepal_width = data.data[:,1]
```

```python
X_sepal_length.shape
```

### Histogram

```python
fig=plt.figure() #Plots in matplotlib reside within a figure object, use plt.figure to create new figure
#Create one or more subplots using add_subplot, because you can't create blank figure
ax = fig.add_subplot(1,1,1)
#Variable
ax.hist(X_sepal_length,bins = 15) # Here you can play with number of bins Labels and Tit
plt.title('Distribution')
plt.xlabel('Sepal')
plt.ylabel('length')
plt.show()
```

### Box Plot

```python
fig=plt.figure()
ax = fig.add_subplot(1,1,1)
#Variable
ax.boxplot(X_sepal_length)
plt.show()
```

### Bar Chart

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
 
objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
y_pos = np.arange(len(objects))
performance = [10,8,6,4,2,1]
 
plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Usage')
plt.title('Programming language usage')
 
plt.show()
```

### Scatter Plot

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(X_sepal_length,X_sepal_width) #You can also add more variables here to represent color and size.
plt.show()
```

### 3D in matplotlib

```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA

X = data.data[:, :2]  # we only take the first two features.
y = data.target

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5

plt.figure(2, figsize=(8, 6))
plt.clf()

# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1,
            edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())

# To getter a better understanding of interaction of the dimensions
# plot the first three PCA dimensions
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)
X_reduced = PCA(n_components=3).fit_transform(data.data)
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=y,
           cmap=plt.cm.Set1, edgecolor='k', s=40)
ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])

plt.show()
```
